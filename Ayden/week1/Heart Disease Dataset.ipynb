```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
heart = pd.read_csv('Data/heart.csv')
heart
```

```python
fig = px.histogram(heart, x="chol")
fig.show()
```

```python
fig = px.scatter(heart, x="age", y="chol")
fig.update_layout(
    title_text = "Scatterplot of Cholesteral and Age")
fig.show()
```

```python
correlation = np.round(heart.corr(), 2)
```

```python
fig = px.imshow(correlation, text_auto = True)
fig.update_layout(
    width = 900,
    height = 900,
    title_text = 'Heat Map of Data' )
fig.show()
```

```python
fig = px.bar(heart, x="age", y="thalach", color="target", facet_row="sex")
fig.show()
```
