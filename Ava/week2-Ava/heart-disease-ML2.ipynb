```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
```

```python
heart = pd.read_csv('hd.csv')
heart
```

```python
heart.dropna(inplace=True)
heart.reset_index(drop=True, inplace=True) #Very good practice to reset how your rows are counted when you drop rows.
```

```python
heart.columns
```

```python
sns.boxplot(x = heart['chol'])
```

```python
heart= heart[heart['ca'] < 2.5]
heart
```

```python
heart= heart[heart['oldpeak'] < 5]
heart
```

```python
heart= heart[heart['thalach'] > 70]
heart
```

```python
heart= heart[heart['chol'] < 380]
heart
```

```python
heart= heart[heart['trestbps'] < 171]
heart
```

```python
correlation = np.round(heart.corr(), 2)
fig= px.imshow(correlation, text_auto = True)
fig.show()
```

```python
x= heart.drop(['target'], axis = 1)
y = heart['target']
```

```python
x, y
```

```python
import sklearn
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.7)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = GaussianNB()
```

```python
clf.fit(x_train, y_train)
```

```python
y_hat = clf.predict(x_test)
```

```python
y_hat
```

```python
from sklearn.metrics import accuracy_score
#find accuracy
accuracy_score(y_hat, y_test)
```

```python
from sklearn.metrics import classification_report
# find classification
classification_report(y_hat, y_test)
```

```python
from sklearn.metrics import recall_score

recall_score(y_hat, y_test)
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, y_hat), annot=True, fmt='g')
```
