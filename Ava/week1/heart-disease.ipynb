column info:

age

sex

cp\- chest pain type \(4 values\)

trestbps\- resting blood pressure

chol\- serum cholestoral in mg/dl

fbs\- fasting blood sugar &gt; 120 mg/dl

thalach\- resting electrocardiographic results \(values 0,1,2\)

exang\- maximum heart rate achieved

oldexercise induced angina

oldpeak = ST depression induced by exercise relative to rest

slope\-the slope of the peak exercise ST segment

ca\- number of major vessels \(0\-3\) colored by flourosopy

thal: 0 = normal; 1 = fixed defect; 2 = reversable defect



```python
heart.dropna(inplace=True)
heart.reset_index(drop=True, inplace=True)
```

```python
len(heart)
```

```python
heart.age.head()
```

```python
fig = px.histogram(heart, x="age")
fig.show()
```

```python
fig = px.scatter(heart, x="trestbps")
fig.show()
```

```python
heart.corr()
```

```python
correlation= heart.corr()
```

```python
fig = px.imshow(correlation, text_auto = True)
fig.update_layout(
    width=900,
    height=900,
    title_text = 'Correlation Heatmap of Data'
)
fig.show()
```

This scatter plot shows the resting blood pressure of a person according to their age. The color of the dot represents their sex, and its size represents their serum cholesterol.



```python
fig = px.scatter(heart, x="trestbps", y="age", color="sex",
                 size='chol')
fig.show()
```

```python

fig = px.scatter(heart, x='age', y='exang', opacity=0.65, trendline='ols', trendline_color_override='darkblue')

fig.show()
```
